Machine Learning is a field of Artificial Intelligence that allows computers to learn patterns from data instead of being explicitly programmed with fixed rules. In traditional programming, developers provide both the data and the instructions, and the computer produces the output. In contrast, in Machine Learning, we provide the data and the correct outputs, and the system automatically learns the rules that map inputs to outputs. After training, the model can use these learned patterns to make predictions on new, unseen data. This makes Machine Learning especially useful when writing detailed rules manually becomes difficult or impractical.

### Day 21 – Types of Machine Learning

1. **Supervised Learning**  
Supervised Learning is a type of Machine Learning where the model is trained on labeled data. Each input has a corresponding correct output. The goal is to learn the mapping between inputs and outputs to make predictions on new data.

Example: Email → Spam or Not Spam.

2. **Unsupervised Learning**  
Unsupervised Learning is used when the data has no labels. The model identifies hidden patterns, structures, or clusters in the dataset.

Example: Grouping customers based on purchasing behavior.

3. **Reinforcement Learning**  
Reinforcement Learning is a method where an agent learns by interacting with an environment and receives rewards or penalties based on actions. The objective is to maximize total reward.

| Learning Type          | Definition                                      | Real-World Example 1        | Real-World Example 2         |
|------------------------|------------------------------------------------|----------------------------|-----------------------------|
| Supervised Learning    | Data has labels (input + correct output)      | Spam Detection              | House Price Prediction      |
| Unsupervised Learning  | Data has no labels; finds hidden patterns     | Customer Segmentation       | Market Basket Analysis      |
| Reinforcement Learning | Learns using rewards and penalties            | Self-Driving Cars           | Game Playing AI             |
